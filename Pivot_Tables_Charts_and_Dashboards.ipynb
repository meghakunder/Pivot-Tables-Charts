**4.Explain the difference between Sorting and Filtering with an example from this dataset.**

  **a). Sorting**

Sorting is used to arrange data in a specific order, such as ascending or descending. It does not hide any rows it only changes the order in which they appear.
Example: If the Sales column is sorted from highest to lowest in this dataset, values like 497, 483, and 467 will appear at the top while lower values move to the bottom.
When sorting this dataset, the option **Expand the selection** must be used.
If only the Sales column is sorted, the values in column D will move independently while the Region, Product, and Quantity columns stay in place. This will misalign the data and make the entire table incorrect.

Expanding the selection ensures that the entire row is sorted together, keeping all fields properly matched.

image.png

**b.)Filtering**

Filtering is used to display only the rows that meet certain criteria while hiding the rest.
Example: If the Region column is filtered to show only West, then only the rows where Region = West will remain visible, and all other region rows will be temporarily hidden.

**7.)Why are Dashboards important for decision-making in business analysis?**

Dashboards are important because they present key business information in a visual and easy-to-understand format. They combine charts, summaries, and performance indicators in one place, allowing users to quickly identify trends, compare metrics, and monitor progress. This helps to  make faster and more informed decisions, as they can immediately see what is performing well and what needs attention. Dashboards also reduce the time spent on manual reporting and provide real-time insights for better strategic planning.
